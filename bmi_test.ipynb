```python
bmi ={ 'UW' :'Underweight',
                    'NW': 'Normal Weight',
                    'OW': 'Over Weight',
                    'MO': 'Moderately obese',
                    'SO': 'Severly obese',
                    'VSO': 'Very Severly obese',
                   }

health= {
                    'MR':'Malnutrition risk',
                    'LR': 'Low risk',
                    'ER': 'Enhanced risk',
                    'MRS': 'Medium risk',
                    'HR': 'High risk',
                    'VHR': 'Very high risk',
    }
```

```python
def calculate_bmi(data):
    bmi_list=[]
    BMI_Category= None
    Health_risk= None
    BMI_range = None
    height =int(data[1])
    mass = int(data[2])
    BMI = mass/((height/100)**2)
    print ("height",height)    
    print ("mass",mass)
    print ('BMI',BMI)
    if BMI <= 18.4 :
        BMI_Category = bmi.get('UW')
        BMI_range = BMI
        Health_risk = health.get('MR')
    elif BMI >=18.5 and BMI <= 24.9:
        BMI_Category = bmi.get('NW')
        BMI_range = BMI
        Health_risk = health.get('LR')
    elif BMI >=25 and BMI <= 29.9:
        BMI_Category = bmi.get("OW")
        BMI_range = BMI
        Health_risk = health.get('ER')
    elif BMI >=30 and BMI <= 34.9:
        BMI_Category = bmi.get("MO")
        BMI_range = BMI
        Health_risk = health.get('MRS')
    elif BMI >= 35 and BMI <= 39.9:
        BMI_Category = bmi.get("SO")
        BMI_range = BMI
        Health_risk = health.get('HR')
    elif BMI > 40:
        BMI_Category= bmi.get("VSO")
        BMI_range = BMI
        Health_risk = health.get('VHR')
    bmi_list.append([BMI_Category,BMI_range,Health_risk])
    return  (bmi_list)

```

# imagine this will be used in-product for 1 million patient in the form of csv

```python
def read_contents():
    list_values = []
    with open('Input.txt') as file:
        lines = csv.reader(file)
        for row in lines:
            list_values.append(row)
        return list_values
```

```python
def csvdatabmi(data):
    line_count = 0
    notValidData = []
    validData = []
    for row in data:
        if line_count == 0:
            line_count = 1
        elif row[0].isalpha() == True and row[1].isnumeric() == True and row[2].isnumeric() == True :
            bmi = calculate_bmi(row)
            validData.append(row+bmi)
            line_count +=1
        else:
            notValidData.append(row)
            line_count +=1
    # print('this data is not valid-->',notValidData)
    # print('total_rows-->',line_count)
    return validData
```

```python
def write_csv(data):
    with open('output.csv', 'w',newline='') as file:
        csv_data = csv.writer(file, delimiter=',')
        csv_data.writerow(['Gender', 'HeightCm', 'WeightKG', 'BMI category', 'BMI Range','Health Risk'])
        for row in data:
            # print(row)
            csv_data.writerows([row])
    return 'added successfully'
```

```python
contents = read_contents()
    # print(result)
finaldata = csvdatabmi(contents)
    # print('final data with BMI-->',finaldata)
result = write_csv(finaldata)
print(result)
```
